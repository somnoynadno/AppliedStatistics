```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**2 - 14

def df(x):
    return 2*x
```

```python
# метод дихотомии
def bissect(f, a, b, eps=1e-9):
    C = []
    while abs(b - a) > eps:
        c = (a + b) / 2

        if f(a) * f(c) > 0:
            a = c
        else:
            b = c
            
        C.append(c)
        
    return np.array(C)    
#     return с
```

```python
# метод хорд
def secant(f, a, b, eps=1e-9):
    C = []
    while abs(b - a) > eps:
        c = a - (b - a)/(f(b) - f(a))*f(a)

        if f(a) * f(c) > 0:
            a = c
        else:
            b = c
            
        C.append(c)
            
    return np.array(C)    
#     return с
```

```python
# метод Ньютона
def newton(f, df, x0, eps=1e-9):
    C = []
    while abs(f(x0)) > eps:
        x0 = x0 - f(x0) / df(x0)
        
        C.append(x0)
        
    return np.array(C)
#     return x0
```

```python
x = np.linspace(0, 5)

plt.plot(x, f(x))
```

```python
x1_ = bissect(f, 3, 4)
x2_ = secant(f, 3, 4)
x3_ = newton(f, df, 3)

err1 = np.abs(x1_ - 14**0.5)
err2 = np.abs(x2_ - 14**0.5)
err3 = np.abs(x3_ - 14**0.5)

plt.semilogy(err1)
plt.semilogy(err2)
plt.semilogy(err3)
```
